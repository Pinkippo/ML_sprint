```python
# 라이브러리 임포트
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#데이터 불러오기
df = pd.read_csv('data3.csv')
```

```python
# 산점도 그리기
plt.scatter(df['date'], df['number_people'], s=1)

# 그래프 제목, 축 라벨 설정
plt.title('Number of People')
plt.xlabel('date')
plt.ylabel('Number of People')

# 그래프 출력
plt.show()
```

```python
# number_people 열의 평균, 최대, 최소 계산
avg_number_people = df['number_people'].mean()
max_number_people = df['number_people'].max()
min_number_people = df['number_people'].min()

# 결과 출력
print('number_people 열의 평균:', avg_number_people)
print('number_people 열의 최대값:', max_number_people)
print('number_people 열의 최소값:', min_number_people)
```
